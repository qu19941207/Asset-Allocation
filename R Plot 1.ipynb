```r
library("readxl")

stock = read_xlsx(".xlsx")
date_stock = do.call(c,stock[1])
stock = matrix(unlist(stock), ncol=6, nrow=224)

plot(date_stock, stock[,2], col = 'blue', type="l", xlab = "Year", ylab = "SSE Composite Index")
plot(date_stock, stock[,3], col = 'blue', type="l", xlab = "Year", ylab = "S&P 500 Index")
plot(date_stock, stock[,4], col = 'blue', type="l", xlab = "Year", ylab = "Nikkei 225")
plot(date_stock, stock[,5], col = 'blue', type="l", xlab = "Year", ylab = "Hang Seng Index")
plot(date_stock, stock[,6], col = 'blue', type="l", xlab = "Year", ylab = "S&P/ASX 200")
```
